# Notebook purpose:
    
This notebook is meant to be run in google colab after a model has been built and saved (see nb: get_protENN_model). This nb will load the test data from the raw data_dir for inference, load in the saved models from the model_dir, and save the classification results to a results_dir.

This is the debugging notebook responsible for inference.py and the inference step of main.py

```python
import os
from collections import Counter

import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive/')

!cp './drive/MyDrive/Colab Notebooks/ProtCNN/utils/datautils.py' ./
from datautils import *

import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
import keras

from keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Add, MaxPooling1D, Flatten, Dropout, Dense, Activation, BatchNormalization, Conv1D
from tensorflow.keras import Model, Input, Sequential
from keras.regularizers import l2

from sklearn.metrics import classification_report
```

```python
data_dir = './drive/MyDrive/PFAM_database/data/random_split/' ##dir in google drive where the raw data lives
train_data, train_targets = reader('train',data_dir)
fam2label = build_labels(train_targets)
word2id = build_vocab(train_data)
model_dir = 'drive/MyDrive/Colab Notebooks/ProtCNN/models' ##dir in google drive where the models live
```

```python
model = tf.keras.models.load_model(model_dir+'/ProtCNN_two_resblock_5_epoch_model.h5') ##call in specific model
```

```python
test = SequenceData(word2id, fam2label, 120, data_dir,"test") ##get test data
test_dict = test.get_data_dictionaries()
```

```python
test_pred = model.predict(np.array(test_dict['sequence'])) ##do prediction
```

```python
test_pred.shape ##verify shape is as expected 
```

```python
y_preds = np.argmax(test_pred, axis = 1) ##get index which is the class label
```

```python
y_true = np.array(test_dict['target']) ##extract the ground truths from the test_data
```

```python
class_rep = classification_report(y_true, y_preds)
print(class_rep)
```

```python
##save full classification report

results_dir = 'drive/MyDrive/Colab Notebooks/ProtCNN/results/'
model_name = 'protCNN_5_epoc_'

# os.mkdir(results_dir)
```

```python
from datetime import datetime
now = datetime.now()
ds_string = now.strftime("%Y_%d_%m_%H:%M:%S_")
print(ds_string)
```

```python
with open(results_dir+model_name+ds_string+"verbose_results.txt", "w") as external_file:
    print(class_rep, file=external_file)
    external_file.close()
```

```python
##save abbreviated classification report
abrev_class_rep = class_rep.split('\n\n')[0]+'\n\n'+class_rep.split('\n\n')[-1]
print(abrev_class_rep)

with open(results_dir+model_name+ds_string+"simp_results.txt", "w") as external_file:
    print(abrev_class_rep, file=external_file)
    external_file.close()
```
